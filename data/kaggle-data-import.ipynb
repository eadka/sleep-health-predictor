```python
import kagglehub
import pandas as pd
import os
import shutil
```

```python
# Download latest version
path = kagglehub.dataset_download("uom190346a/sleep-health-and-lifestyle-dataset")

print("Path to dataset files:", path)
```

#### Lets save the datafile into the data folder

```python
# Defining the source and destination paths
src_file = os.path.join(path, "Sleep_health_and_lifestyle_dataset.csv")
dst_file = os.path.join("./sleep_health_lifestyle.csv")
```

```python
# Copy the file to the current folder
shutil.copy(src_file, dst_file)
```
